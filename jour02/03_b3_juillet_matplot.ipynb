```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fleurs = load_iris()

X = fleurs.data
Y = fleurs.target
ax = plt.axes(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2] , c=Y)
```

```python
f = lambda x, y: np.sin(x) + np.cos(x+y)

X = np.linspace(0,5, 1000)
Y = np.linspace(0,5, 1000)
X,Y = np.meshgrid(X, Y )
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z , cmap="plasma")
```

```python
# j'ai besoin de créer un histogramme
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fleurs = load_iris()

X = fleurs.data
Y = fleurs.target

plt.figure()

plt.subplot(2,2,1)

plt.hist(X[:,0], color="yellow")
plt.title(f"Répartition {fleurs["feature_names"][0]}")

plt.subplot(2,2,2)

plt.hist(X[:,1], color="orange")
plt.title(f"Répartition {fleurs["feature_names"][1]}")

plt.subplot(2,2,3)

plt.scatter(X[:,2] , X[:,3] , c="r")
plt.title(f"Nuage de pt {fleurs["feature_names"][2]}")

plt.subplot(2,2,4)

plt.scatter(X[:,3] , X[:,2] , c="g")
plt.title(f"Nuage de pt {fleurs["feature_names"][3]}")

plt.tight_layout()
```

```python
# histogramme en 2d
# réussir à présenter 3 informations avec un graphique en 2dimensions (hist2d)


plt.hist2d( X[:,0] , X[:,1] , cmap="Blues")
plt.xlabel("longueur sépale")
plt.ylabel("largueur sépale")
plt.colorbar()
```

```python
# contour plt => un graphique de type carte de montage


f = lambda x, y: np.sin(x) + np.cos(x+y)

X = np.linspace(0,5, 1000)
Y = np.linspace(0,5, 1000)
X,Y = np.meshgrid(X, Y )
Z = f(X, Y)

plt.contour( X,Y,Z ,20 , cmap="RdGy")
plt.colorbar()
```
